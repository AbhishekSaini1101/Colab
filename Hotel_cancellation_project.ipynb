```python
#Data Analysis Project
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

Loading Dataset

```python
dataset = pd.read_csv('/content/hotel_booking.csv')

dataset.head()
```

```python
dataset.tail()
```

```python
dataset.shape
```

```python
 dataset.info()
```

```python
#Date is given in Object format
#Converting Date into Date Format

dataset['reservation_status_date'] = pd.to_datetime(dataset['reservation_status_date'])
```

```python
dataset.info()
```

```python
dataset.describe(include = 'object')
```

```python
#For loop to get all the columns of Object type ( dataset.describe(include='object'))

for col in dataset.describe(include ='object').columns:
  print(col)
  print(dataset[col].unique())
  print("-"*50)
```

```python
dataset.isnull().sum()
```

```python
#Removing null value Column

dataset.drop(["company","agent"], axis=1, inplace=True)
dataset.dropna(inplace = True)
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
#Now we have cleaned our data

dataset = dataset[dataset["adr"]<5000]
```

Data Analysis and Visualization
on cleaned Data

```python
#value_count returns the Category as well as count the number of values

cancelled_perc = dataset["is_canceled"].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize = (5,4))
plt.title('Reservation Status Count')
plt.bar(['Not canceled','Canceled'], dataset["is_canceled"].value_counts(), edgecolor='k', width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'canceled', data = dataset, palette = 'Blues')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```
